```python
def evolution_walk(tree : dict):
    for evolution in tree['evolves_to']:
        evolution_names[tree['species']['name'] + str(random.randint(0, 100000000000000000000))] = evolution_walk(evolution)
    if tree['species']['name']:
        try:
            if tree['evolution_details'][0]['trigger']['name'] == 'trade':
                return [tree['species']['name'], tree['evolution_details'][0]['trigger']['name']]

            if tree['evolution_details'][0]['trigger']['name'] == 'level-up':

                return [tree['species']['name'], tree['evolution_details'][0]['trigger']['name'], tree['evolution_details'][0]['min_level']]
            if tree['evolution_details'][0]['trigger']['name'] == 'use-item':
                return [tree['species']['name'], tree['evolution_details'][0]['trigger']['name'], tree['evolution_details'][0]['item']['name']]

        except:
            return tree['species']['name']


def get_evolution_chain(tree : dict):
    tree = tree['chain']
    evolution_walk(tree)

```

```python
def str_start_up(string : str):
    return string[0].upper()+string[1:].lower()

def evolution_walk(evolution):
    if isinstance(evolution, str):
        evolution = ast.literal_eval(evolution)

    if evolution:
        if isinstance(evolution[0], list):
            next_evolution = df_pokemon.loc[df_pokemon.index[df_pokemon['Name'] == str_start_up(evolution[0][0])][0], "evolution"]
        else:
            next_evolution = df_pokemon.loc[df_pokemon.index[df_pokemon['Name'] == str_start_up(evolution[0])][0], "evolution"]

        evolution_walk(next_evolution)
    if not evolution:
        return " "
    if evolution[1] == 'level-up':
        return f'  --- Lvl. {evolution[2]} ---> {evolution[0]}'
    if evolution[1] == 'use-item':
        return f'---  {evolution[2][0].upper()}{evolution[2][1:]} ---> {evolution[0]}'
    if evolution[1] == "  --- trade ---> ":
        return f'  --- Trade ---> {evolution[2]}'




def print_evolution_order(df_current_pokemon):
    return_string = ""
    evolution_possibilities = ['level-up', 'use-item', 'trade']
    evolutions = ast.literal_eval(df_current_pokemon.loc['evolution'])
    for x in evolutions:
        return_string = return_string + f"\n\n{df_current_pokemon.loc['Name']}"
        return_string = return_string + evolution_walk(x)



    return return_string

```

```python
import pyautogui
import win32gui

def screenshot(window_title=None):
    if window_title:
        hwnd = win32gui.FindWindow(None, window_title)
        if hwnd:
            win32gui.SetForegroundWindow(hwnd)
            x, y, x1, y1 = win32gui.GetClientRect(hwnd)
            x, y = win32gui.ClientToScreen(hwnd, (x, y))
            x1, y1 = win32gui.ClientToScreen(hwnd, (x1 - x, y1 - y))
            im = pyautogui.screenshot(region=(x, y, x1, y1))
            return im
        else:
            print('Window not found!')
    else:
        im = pyautogui.screenshot()
        return im
def winEnumHandler( hwnd, ctx ):
    if win32gui.IsWindowVisible( hwnd ):
        print (win32gui.GetWindowText( hwnd ))

win32gui.EnumWindows( winEnumHandler, None )

im = screenshot('Whatsapp')
if im:
    im.show()
```
